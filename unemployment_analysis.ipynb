```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
dataset=pd.read_csv(r'/content/Unemployment_n_India.csv')
print('The shape of the dataset is:',dataset.shape)
print(dataset.head())
print(dataset.info())
print(dataset.describe())
```

```python
dataset.columns= ["Region","Date","Frequency",
               "Estimated Unemployment Rate(%)","Estimated Employed",
               "Estimated Labour Participation Rate(%)","Area"]
plt.title("Unemployment in India")
sns.histplot(x="Estimated Employed", hue="Area", data=dataset)
plt.show()
```

```python
plt.title("Unemployment in India per year")
sns.histplot(x="Estimated Employed", hue="Date", data=dataset)
plt.show()
```

```python
sns.lmplot(x="Estimated Employed", y="Estimated Unemployment Rate(%)", data=dataset, hue="Area", fit_reg=False, legend=False)
plt.legend()
plt.show()
```

```python
fig = px.box(dataset,x='Area',y='Estimated Unemployment Rate(%)',color='Region',title='Box-plot:Unemployment rate',template='plotly')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```
